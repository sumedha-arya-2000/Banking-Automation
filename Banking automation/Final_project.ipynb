```python
from tkinter import *
from PIL import Image,ImageTk
import time
from tkinter import messagebox
import sqlite3
con=sqlite3.connect(database="banking.sqlite")
cur=con.cursor()

table1="""create table account
(acn integer primary key autoincrement,name text,pass text,
mob text,email text,bal float,acnopendate text,acntype text)"""

table2="""create table txn(acn integer,txndate text,txntype text,amt float,
updatedbal float)"""

try:
    cur.execute(table1)
    cur.execute(table2)
    con.commit()
    print("tables created")
except Exception as e:
    print(e)

win=Tk()
win.state("zoomed")
win.configure(bg="powder blue")

title=Label(win,fg='blue',text="Banking Automation",bg="powder blue",font=('Arial',50,'bold','underline'))
title.pack()

img=Image.open("logo.jpg")
img=img.resize((200,100))
imgtk=ImageTk.PhotoImage(img,master=win)

logolbl=Label(win,image=imgtk)
logolbl.place(relx=0,rely=0)

def home_screen():
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg='pink')
    
    def openaccount():
        frm.destroy()
        openaccount_screen()
     
    def forgot():
        frm.destroy()
        forgot_screen()
    
    def login():
        acn=acnentry.get()
        pwd=passentry.get()
        if(len(acn)==0 or len(pwd)==0):
            messagebox.showwarning("validation","Fields can't be empty")
            return
        else:
            con=sqlite3.connect(database="banking.sqlite")
            cur=con.cursor()
            query="select * from account where acn=? and pass=?"
            cur.execute(query,(acn,pwd))
            global usertup
            usertup=cur.fetchone()
            print(usertup)
            if(usertup==None):
                messagebox.showerror("Failed","Invalid ACN/Pass")
                return
            else:
                frm.destroy()
                welcome_frame()
    
        
    acnlbl=Label(frm,text="ACN:",bg='pink',font=('Arial',20,'bold'))
    acnlbl.place(relx=.3,rely=.2)
    
    acnentry=Entry(frm,font=('Arial',20,'bold'),bd=5)
    acnentry.place(relx=.4,rely=.2)
    acnentry.focus()

    passlbl=Label(frm,text="Pass:",bg='pink',font=('Arial',20,'bold'))
    passlbl.place(relx=.3,rely=.3)
    
    passentry=Entry(frm,font=('Arial',20,'bold'),bd=5,show="*")
    passentry.place(relx=.4,rely=.3)

    loginbtn=Button(frm,command=login,font=('Arial',20,'bold'),bd=5,text="login")
    loginbtn.place(relx=.4,rely=.45)
    
    resetbtn=Button(frm,font=('Arial',20,'bold'),bd=5,text="reset")
    resetbtn.place(relx=.5,rely=.45)
    
    fpbtn=Button(frm,command=forgot,width=16,font=('Arial',20,'bold'),bd=5,text="forgot password")
    fpbtn.place(relx=.37,rely=.6)
    
    openbtn=Button(frm,command=openaccount,width=20,font=('Arial',20,'bold'),bd=5,text="open new account")
    openbtn.place(relx=.34,rely=.75)
    
    
def openaccount_screen():
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg='pink')
    
    def back():
        frm.destroy()
        home_screen()
    
    def reset():
        nameentry.delete(0,"end")
        passentry.delete(0,"end")
        emailentry.delete(0,"end")
        mobentry.delete(0,"end")
        nameentry.focus()
    
    def openacn():
        name=nameentry.get()
        pwd=passentry.get()
        mob=mobentry.get()
        email=emailentry.get()
        acntype="Saving"
        opendate=time.ctime()
        bal=0
        if(len(name)==0 or len(pwd)==0 or len(mob)==0 or len(email)==0):
            messagebox.showwarning("validation","Fields can't be empty")
            return
        else:
            con=sqlite3.connect(database="banking.sqlite")
            cur=con.cursor()
            query="""insert into account(name,pass,mob,email,bal,acnopendate,acntype)
            values(?,?,?,?,?,?,?)"""
            cur.execute(query,(name,pwd,mob,email,bal,opendate,acntype))
            con.commit()
            
            cur=con.cursor()
            cur.execute("select max(acn) from account")
            tup=cur.fetchone()
            con.close()
            messagebox.showinfo("Success",f"Your Account No is:{tup[0]}")
            frm.destroy()
            home_screen()
        
    backbtn=Button(frm,command=back,font=('Arial',20,'bold'),bd=5,text="back")
    backbtn.place(relx=0,rely=0)
    
    frmtitle=Label(frm,text="Open New Account",font=('Arial',25,'bold','underline'),bg='pink',fg='red')
    frmtitle.pack()
    
    namelbl=Label(frm,text="Name:",bg='pink',font=('Arial',20,'bold'))
    namelbl.place(relx=.3,rely=.2)
    
    nameentry=Entry(frm,font=('Arial',20,'bold'),bd=5)
    nameentry.place(relx=.4,rely=.2)

    passlbl=Label(frm,text="Pass:",bg='pink',font=('Arial',20,'bold'))
    passlbl.place(relx=.3,rely=.3)
    
    passentry=Entry(frm,font=('Arial',20,'bold'),bd=5,show="*")
    passentry.place(relx=.4,rely=.3)
    
    moblbl=Label(frm,text="Mob:",bg='pink',font=('Arial',20,'bold'))
    moblbl.place(relx=.3,rely=.4)
    
    mobentry=Entry(frm,font=('Arial',20,'bold'),bd=5)
    mobentry.place(relx=.4,rely=.4)

    emaillbl=Label(frm,text="Email:",bg='pink',font=('Arial',20,'bold'))
    emaillbl.place(relx=.3,rely=.5)
    
    emailentry=Entry(frm,font=('Arial',20,'bold'),bd=5)
    emailentry.place(relx=.4,rely=.5)
    
    openbtn=Button(frm,command=openacn,font=('Arial',20,'bold'),bd=5,text="open")
    openbtn.place(relx=.4,rely=.7)
    
    resetbtn=Button(frm,command=reset,font=('Arial',20,'bold'),bd=5,text="reset")
    resetbtn.place(relx=.5,rely=.7)

    
    
def forgot_screen():
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg='pink')
    
    def back():
        frm.destroy()
        home_screen()
        
    def get():
        a=acnentry.get()
        m=mobentry.get()
        e=emailentry.get()
        con=sqlite3.connect(database="banking.sqlite")
        cur=con.cursor()
        query="select pass from account where acn=? and mob=? and email=?"
        cur.execute(query,(a,m,e))
        tup=cur.fetchone()
        if(tup==None):
            messagebox.showerror("Incorrect","Details not found")
        else:
            messagebox.showinfo("Success",f"Your pass is:{tup[0]}")
            frm.destroy()
            home_screen()
    backbtn=Button(frm,command=back,font=('Arial',20,'bold'),bd=5,text="back")
    backbtn.place(relx=0,rely=0)
    
    frmtitle=Label(frm,text="Get Your Password",font=('Arial',25,'bold','underline'),bg='pink',fg='red')
    frmtitle.pack()
    
    acnlbl=Label(frm,text="ACN:",bg='pink',font=('Arial',20,'bold'))
    acnlbl.place(relx=.3,rely=.3)
    
    acnentry=Entry(frm,font=('Arial',20,'bold'),bd=5)
    acnentry.place(relx=.4,rely=.3)
    acnentry.focus()
    
    moblbl=Label(frm,text="Mob:",bg='pink',font=('Arial',20,'bold'))
    moblbl.place(relx=.3,rely=.4)
    
    mobentry=Entry(frm,font=('Arial',20,'bold'),bd=5)
    mobentry.place(relx=.4,rely=.4)

    emaillbl=Label(frm,text="Email:",bg='pink',font=('Arial',20,'bold'))
    emaillbl.place(relx=.3,rely=.5)
    
    emailentry=Entry(frm,font=('Arial',20,'bold'),bd=5)
    emailentry.place(relx=.4,rely=.5)
    
    getbtn=Button(frm,command=get,font=('Arial',20,'bold'),bd=5,text="get")
    getbtn.place(relx=.4,rely=.7)
    
    resetbtn=Button(frm,font=('Arial',20,'bold'),bd=5,text="reset")
    resetbtn.place(relx=.5,rely=.7)

    
    
def welcome_frame():
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg='pink')
    
    def logout():
        frm.destroy()
        home_screen()
        
    def details():
        nestedfrm=Frame(frm,highlightthickness=3,highlightbackground='brown')
        nestedfrm.configure(bg="powder blue")
        nestedfrm.place(relx=.25,rely=.2,relwidth=.6,relheight=.6)
        
        frmtitle=Label(nestedfrm,text="Your Details",font=('Arial',25,'bold','underline'),bg='powder blue',fg='red')
        frmtitle.pack()
        
        con=sqlite3.connect(database="banking.sqlite")
        cur=con.cursor()
        query="select * from account where acn=?"
        cur.execute(query,(usertup[0],))
        tup=cur.fetchone()
        con.close()
        la=Label(nestedfrm,text=f"Account Number: {tup[0]}",bg='powder blue',font=('Arial',20))
        la.place(relx=.1,rely=.2)
        
        lb=Label(nestedfrm,text=f"Current Balance: {tup[5]}",bg='powder blue',font=('Arial',20))
        lb.place(relx=.1,rely=.3)
        
        ldate=Label(nestedfrm,text=f"Account Open Date: {tup[6]}",bg='powder blue',font=('Arial',20))
        ldate.place(relx=.1,rely=.4)
        
        
    def updateprofile():
        nestedfrm=Frame(frm,highlightthickness=3,highlightbackground='brown')
        nestedfrm.configure(bg="powder blue")
        nestedfrm.place(relx=.25,rely=.2,relwidth=.6,relheight=.6)
        
        def changepass():
            np=lpe.get()
            con=sqlite3.connect(database="banking.sqlite")
            cur=con.cursor()
            cur.execute("update account set pass=? where acn=?",(np,usertup[0]))
            con.commit()
            con.close()
            messagebox.showinfo("Success","Password updated")
            lpe.delete(0,"end")
            return
        frmtitle=Label(nestedfrm,text="Update Your Password",font=('Arial',25,'bold','underline'),bg='powder blue',fg='red')
        frmtitle.pack()
    
        lp=Label(nestedfrm,text="Enter New Pass:",bg='powder blue',font=('Arial',15))
        lp.place(relx=.1,rely=.2)
        
        lpe=Entry(nestedfrm,font=('Arial',20),bd=5)
        lpe.place(relx=.35,rely=.2)
        
        lpb=Button(nestedfrm,font=('Arial',20),bd=5,text='update',command=changepass)
        lpb.place(relx=.4,rely=.4)
        
        
    
    def deposit():
        nestedfrm=Frame(frm,highlightthickness=3,highlightbackground='brown')
        nestedfrm.configure(bg="powder blue")
        nestedfrm.place(relx=.25,rely=.2,relwidth=.6,relheight=.6)
        
        def updatebal():
            amt=float(dpentry.get())
            con=sqlite3.connect(database="banking.sqlite")
            cur=con.cursor()
            query="update account set bal=bal+? where acn=?"
            cur.execute(query,(amt,usertup[0]))
            con.commit()
            
            cur=con.cursor()
            cur.execute("select bal from account where acn=?",(usertup[0],))
            bal=cur.fetchone()[0]
            
            cur=con.cursor()
            query="insert into txn values(?,?,?,?,?)"
            cur.execute(query,(usertup[0],time.ctime(),"Cr.",amt,bal))
            con.commit()
            con.close()
            
            messagebox.showinfo("Success","Amount deposited")
            return
            
        frmtitle=Label(nestedfrm,text="Deposit To Your Account",font=('Arial',25,'bold','underline'),bg='powder blue',fg='red')
        frmtitle.pack()
        
        dplbl=Label(nestedfrm,text="Amount:",font=('',20,'bold'),bg='powder blue')
        dplbl.place(relx=.2,rely=.3)
        
        dpentry=Entry(nestedfrm,font=('',15,'bold'),bd=5)
        dpentry.place(relx=.4,rely=.3)
        
        dpbtn=Button(nestedfrm,text="deposit",font=('',15,'bold'),bd=5,command=updatebal)
        dpbtn.place(relx=.3,rely=.5)
        
    
    def withdraw():
        nestedfrm=Frame(frm,highlightthickness=3,highlightbackground='brown')
        nestedfrm.configure(bg="powder blue")
        nestedfrm.place(relx=.25,rely=.2,relwidth=.6,relheight=.6)
        
        frmtitle=Label(nestedfrm,text="Withdraw From Your Account",font=('Arial',25,'bold','underline'),bg='powder blue',fg='red')
        frmtitle.pack()
    
        
        def updatebal():
            amt=float(wtentry.get())
            con=sqlite3.connect(database="banking.sqlite")
            cur=con.cursor()
            
            cur=con.cursor()
            cur.execute("select bal from account where acn=?",(usertup[0],))
            bal=cur.fetchone()[0]
            
            if(bal>=amt):
                query="update account set bal=bal-? where acn=?"
                cur.execute(query,(amt,usertup[0]))
                con.commit()

                cur=con.cursor()
                query="insert into txn values(?,?,?,?,?)"
                cur.execute(query,(usertup[0],time.ctime(),"Dr.",amt,bal-amt))
                con.commit()
                con.close()

                messagebox.showinfo("Success","Amount withdrawn")
                return
            else:
                messagebox.showwarning("Error","Insufficient bal") 
    
        wtlbl=Label(nestedfrm,text="Amount:",font=('',20,'bold'),bg='powder blue')
        wtlbl.place(relx=.2,rely=.3)
        
        wtentry=Entry(nestedfrm,font=('',15,'bold'),bd=5)
        wtentry.place(relx=.4,rely=.3)
        
        wtbtn=Button(nestedfrm,text="withdraw",font=('',15,'bold'),bd=5,command=updatebal)
        wtbtn.place(relx=.3,rely=.5)
    
    def txnhistory():
        nestedfrm=Frame(frm,highlightthickness=3,highlightbackground='brown')
        nestedfrm.configure(bg="powder blue")
        nestedfrm.place(relx=.25,rely=.2,relwidth=.6,relheight=.6)
        
        frmtitle=Label(nestedfrm,text="Your Txn History",font=('Arial',25,'bold','underline'),bg='powder blue',fg='red')
        frmtitle.pack()
    
        con=sqlite3.connect(database="banking.sqlite")
        cur=con.cursor()
        query="select * from txn where acn=?"
        cur.execute(query,(usertup[0],))
        tups=cur.fetchall()
        
        head=Label(nestedfrm,text="\tDate\t\tAmt\tType\tBal",bg='powder blue',font=('Arial',15,'bold','underline'))
        head.place(relx=.1,rely=.2)
        
        i=.3
        for tup in tups:
            row=Label(nestedfrm,text=f"{tup[1]}\t\t{tup[3]}\t{tup[2]}\t{tup[4]}",bg='powder blue',font=('Arial',12))
            row.place(relx=.1,rely=i)
            i=i+.1
       
        
    logoutbtn=Button(frm,command=logout,font=('Arial',20,'bold'),bd=5,text="logout")
    logoutbtn.place(relx=.9,rely=0)
    
    wellbl=Label(frm,text=f"Welcome,{usertup[1]}",bg='pink',font=('Arial',15,'bold'),fg='green')
    wellbl.place(relx=0,rely=0)
    
    timelbl=Label(frm,text="login time:"+time.ctime(),bg='pink',font=('Arial',15,'bold'))
    timelbl.pack()
    
    detailsbtn=Button(frm,command=details,font=('Arial',20,'bold'),bd=5,text="details",width=12)
    detailsbtn.place(relx=0,rely=.12)
    
    updatebtn=Button(frm,command=updateprofile,font=('Arial',20,'bold'),bd=5,text="update pass",width=12)
    updatebtn.place(relx=0,rely=.25)
    
    depbtn=Button(frm,command=deposit,font=('Arial',20,'bold'),bd=5,text="deposit",width=12)
    depbtn.place(relx=0,rely=.38)
    
    withbtn=Button(frm,command=withdraw,font=('Arial',20,'bold'),bd=5,text="withdraw",width=12)
    withbtn.place(relx=0,rely=.51)
    
    txnbtn=Button(frm,command=txnhistory,font=('Arial',20,'bold'),bd=5,text="txn history",width=12)
    txnbtn.place(relx=0,rely=.64)
    
    
home_screen()
win.mainloop()

```
